```python
trends_df_agg = trends_df.groupby(["date", "display_term"], as_index=False).sum()
```

```python
import plotly.graph_objects as go
from plotly.colors import n_colors
import numpy as np
import plotly.express as px

fig = px.line(trends_df_agg, x='date', y='orig_score_difference',
                facet_col='display_term', facet_col_wrap=3)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.update_yaxes(title=None, showticklabels=False)
fig.update_xaxes(title=None)
fig.update_traces(hovertemplate="<b>baking</b><br>%{x}<br>Score Difference from Previous Year: %{y}<extra></extra>")
fig.show()


print("plotly express hovertemplate:", fig.data[0].hovertemplate)
```
